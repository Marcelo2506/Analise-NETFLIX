```python
netflix = df.copy() #Criando uma cópia para que possamos implementar formato data
```

```python
netflix["date_added"] = pd.to_datetime(netflix['date_added'])
netflix['day_added'] = netflix['date_added'].dt.day
netflix['year_added'] = netflix['date_added'].dt.year
netflix['month_added']=netflix['date_added'].dt.month
netflix['year_added'].astype(int);
netflix['day_added'].astype(int);
```

```python
netflix.head(5)
```

# VISUALIZAÇÃO DOS DADOS

```python
nt = netflix['type']
nt.value_counts().plot(kind='bar')
nt.value_counts()
```

```python
#EM GRÁFICO PIZZA
labels = ['Movie','Tv show']
nt.value_counts().plot(kind='pie')
plt.legend()
```

```python
#RANKING DOS TIPOS DE CONTEÚDOS GERADOS
netflix['rating'].value_counts().plot(kind='bar')
plt.title('Rating')
```

```python
#PUDEMOS CONFERIR NO GRÁFICO ACIMA QUE A TV-MA(Audiência madura — não recomendado para menores de 17 anos) LIDERA O RANKING
```

```python
#RELAÇÃO ENTRE TIPO E CLASSIFICAÇÃO
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relação entre tipo e classificação')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
#AQUI PODEMOS VER OS PAÍSES MAIS PRODUZEM CONTEÚDO
wordcloud = WordCloud(background_color="black",
                      width=1920, height=1080,
                      min_font_size=1).generate(" ".join(netflix['country']))

# mostrar a imagem final
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('country.png')
plt.show()
```

```python
#NOMES DOS DIRETORES MAIS FREQUENTES
wordcloud = WordCloud(background_color="black",
                      width=1920, height=1080,
                      min_font_size=1).generate(" ".join(netflix['director']))

# mostrar a imagem final
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('director.png')
plt.show()
```

```python
#TIPOS DE CONTEUDO MAIS PRODUZIDOS
wordcloud = WordCloud(background_color="black",
                      width=1920, height=1080,
                      min_font_size=1).generate(" ".join(netflix['listed_in']))

# mostrar a imagem final
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('listed.png')
plt.show()
```
